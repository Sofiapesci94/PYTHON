```python
# • quante attività ci sono nel quadrato di longitudine 9-10 e latitudine 45-46?

longitudine = (pub['Longitudine'] >= 9) & (pub['Longitudine'] <=10)                 #creo filtro per vedere quanti TRUE ci sono che corrispondono alle condizioni che ho dato
latitudine = (pub['Latitudine'] >= 45) & (pub['Latitudine'] <=46)         

long_lat = pub.loc[longitudine & latitudine]                                        #mi creo il filtro per la tabella danto valori numerici o nominativi

print(long_lat)
print(long_lat.shape)                                                               #stampo!
```

```python
boston = pd.read_csv('DATABASES/boston.csv')
boston


# La media del prezzo delle case cambia a seconda della distanza dal fiume Charles? 
# • Si nota una correlazione tra il tasso di criminalità e il valore delle abitazioni? 
# Come si può spiegare il risultato?
# • Qual è la media del numero di stanze rispetto al rapporto alunni-insegnanti? 
# E del valore delle case? Appare esserci una qualche correlazione? 
# Come si può spiegare il risultato? 
# • Rispetto all'accessibilità alle autostrade, cambia qualcosa la media delle età delle abitazioni?
# E del numero di stanze? E delle tasse?
```

```python
# • CRIM -tasso di criminalità pro capite per città 
# • ZN - proporzione di terreni residenziali suddivisi in zone per lotti superiori a 25.000 piedi quadrati.
# • INDUS - percentuale di acri di attività commerciali non al dettaglio per città. 
# • CHAS -variabile dummy del fiume Charles (1 se il tratto costeggia il fiume; 0 altrimenti)
# • NOX -concentrazione di ossidi di azoto (parti per 10 milioni).
# • RM -numero medio di stanze per abitazione 
# • AGE -proporzione di unità abitative occupate da proprietari costruite prima del 1940 
# • DIS - distanze ponderate da cinque centri occupazionali di Boston 
# • RAD -indice di accessibilità alle autostrade radiali 
# • TAX -aliquota dell'imposta fondiaria sul valore pieno per 10.000 dollari 
# • PTRATIO - rapporto alunni-insegnanti per città 
# • BLACK -la percentuale di neri per città 
# • LSTAT - % di popolazione di condizione più bassa 8 
# • MEDV -Valore mediano delle case, espresso in migliaia di dollari
```

```python
# La media del prezzo delle case cambia a seconda della distanza dal fiume Charles? 
boston.groupby("chas")["medv"].mean()
```

```python
# • Si nota una correlazione tra il tasso di criminalità e il valore delle abitazioni? 
# CRIM MEDV
boston.groupby("crim")["medv"].mean().iloc[-20:]
```

```python
# • Qual è la media del numero di stanze rispetto al rapporto alunni-insegnanti? 
# RM PTRATIO
boston.groupby('rm')['ptratio'].mean()
```

```python
# E del valore delle case? Appare esserci una qualche correlazione? Come si può spiegare il risultato?

ptratio_vs_medv = boston.groupby("ptratio")["medv"].mean()

ptratio_vs_medv = ptratio_vs_medv.reset_index()
# Il metodo reset_index() serve a riportare l'indice della Series a una colonna normale del DataFrame, 
# e a ripristinare l'indice numerico predefinito (0, 1, 2, ...). 
# In pratica, trasforma la Series in un DataFrame, mantenendo i dati come colonne separate.

    
ptratio_vs_medv.plot(kind="scatter", x="ptratio", y="medv")
# Questo comando crea un grafico a dispersione (scatter plot) utilizzando il DataFrame ptratio_vs_medv.
# kind="scatter": Specifica che il tipo di grafico che vogliamo creare è un grafico a dispersione, dove i punti sono tracciati su un piano cartesiano.
# x="ptratio": La colonna ptratio verrà usata per i valori sull'asse x del grafico.
# y="medv": La colonna medv verrà usata per i valori sull'asse y del grafico.
# Quindi, questo comando crea un grafico che mostra come la media del valore delle case (medv) varia in funzione del rapporto studenti/insegnante (ptratio).
```
