```python
# Se prendiamo un quarto di tutti i prodotti, in modo che siano i più costosi, quale sarà il loro range di prezzo?

prezzo = (dimproduct['ListPrice'].dropna().sort_values())             #metto la lista prezzi in ordine crescente  DROPNA TOGLIE I VALORI NULLI
print(prezzo)


calcolo_quarti = prezzo.count()                     #conto tutti i pezzi che ho nella lista senza null per dividerli in quarti
risultato = calcolo_quarti/4                        #divido la lista prezzi in 4/4  
print('\n1/4=',round(risultato))                    #mi calcolo l'ultimo quarto

media = prezzo.tail(99).mean()                                        #tail - prendo gli ultimi 90 valori perché

print('\nla media del quarto dei prezzi più costosi è:', round(media))    


```

```python
# PREZZO MEDIO PER I PRODOTTI BLU

filtro_colore = dimproduct.loc[:,'Color']== 'Blue'
# Qui stiamo cercando tutti i prodotti nella colonna 'Color' che hanno il valore 'Blue'.
# Il risultato è una lista di True (vero) e False (falso), dove True indica che il prodotto è blu e False che non lo è.

dimproduct_color = dimproduct.loc[filtro_colore]                   
# Creare una nuova tabella con solo i prodotti blu
# Usiamo la lista di True e False per selezionare solo le righe della tabella originale (dimproduct) dove il colore è blu.
# Il risultato è una nuova tabella (dimproduct_color) che contiene solo i prodotti blu

prezzo_medio = dimproduct_color['DealerPrice'].mean().round(2)  
#creo la variabile del prezzo medio, dopo essermi creata la colonna di solo prodotti blu la affianco a dealer price e calcolo la media arrotondata

print('il prezzo medio dei prodotti blu è:', prezzo_medio)
```

```python
#PREZZO MEDIO PRODOTTI ROSSI E NERI

#creiamo due filtri
filtro_rosso = dimproduct.loc[:, 'Color'] == 'Red'      
filtro_nero = dimproduct.loc[:, 'Color'] == 'Black'
# Cerca tutti i prodotti nella colonna 'Color' che corrispondono (==) a 'Red' e 'Black' che 
# Il risultato è una lista di True (vero) e False (falso), dove True indica che il prodotto è rosso e False che non lo è.

filtro_colore = filtro_rosso | filtro_nero
# Combina i due filtri (rosso_nero) usando l'operatore OR (|), che significa "oppure".
# Se un prodotto è rosso OPPURE nero, il risultato sarà True.
# Se un prodotto non è né rosso né nero, il risultato sarà False.

    
prezzo_medio = dimproduct.loc[filtro_colore, 'DealerPrice'].mean().round(2)
# mi creo una tabella (.loc) contenente[filtro , DP] calcolo la media(.mean) e arrotondo a 2(round)

print('il prezzo medio dei prodotti rossi e neri sarà di:', prezzo_medio)
```

```python
# PREZZO MASSIMO PER PRODOTTI TAGLIA 42 E PESO OLTRE 10 Kg color argento

# STAMPIAMO COLONNE 
dimproduct.columns

#filtri
filtro_42 = dimproduct['Size'] == '42'
filtro_peso = dimproduct['Weight'] >10
filtro_col = dimproduct['Color'] == 'Silver'
# Cerca tutti i prodotti nella colonne [Size, Weight, Color] che corrispondono (==) o siano > di quel che ci serve sapere
# Il risultato è una lista di True (vero) e False (falso), dove True indica che il prodotto con la caratteristica ricercata e False il prodotto che non la ha.

filtro_row = filtro_42 & filtro_peso & filtro_col
# Combina le tre maschere booleane (filtri 42,peso, col) usando l'operatore AND (&), che significa "e".
# Un valore sarà True solo se il prodotto soddisfa tutte e tre le condizioni: taglia 42, peso > 10 kg e colore argento. (le uniche che combaciano in tutto)
# Un valore sarà False se il prodotto non soddisfa almeno una delle condizioni.

filtro_col = ['EnglishDescription', 'StandardCost']
#crea una lista (tabella) con le colonne che voglio visualizzare

display(dimproduct.loc[filtro_row, filtro_col])
# Filtra il DataFrame dimproduct usando la maschera filtro_row. 
# Ottieni una nuova tabella che contiene solo i prodotti che soddisfano tutte e tre le condizioni. (le evolve dal true false e mette in lome prodotto)
# Dalla tabella filtrata, seleziona solo le colonne 'EnglishDescription' e 'StandardCost'.
```

```python
#DATASET DIABETE

diabetes = pd.read_csv('DATABASES/diabetes.csv')
diabetes

# • Osserviamone le dimensioni e un'anteprima di cinque righe; 
# • Prendiamoci un po' di tempo per dare un'occhiata ai metadati delle colonne; 
# • Stampiamo dei descrittori statistici del dataset; 
# • Selezioniamo i dati relativi a diverse fasce di età: <20, 20-30, 30-40, 40-50, >50; 
# • Qual è la media della pressione sanguigna diastolica per le diverse fasce di età? 6 
# • Qual è la media della pressione per ogni singolo anno di età?


# Osserviamone le dimensioni e un'anteprima di cinque righe; 
diabetes.shape
diabetes.head(5)


# • Prendiamoci un po' di tempo per dare un'occhiata ai metadati delle colonne; 
diabetes.columns


# • Stampiamo dei descrittori statistici del dataset; 
diabetes.describe()


# • Selezioniamo i dati relativi a diverse fasce di età: <20, 20-30, 30-40, 40-50, >50; 
Filtro_minore20 = diabetes['Age (years)'] <20                                            #filtro per i dati minori di 20       (True/False)
Filtro_compreso1 = (diabetes['Age (years)']>=20) & (diabetes['Age (years)']<=30)         #filtro per i maggiori= a 20 e minori = a 30
Filtro_compreso2 = (diabetes['Age (years)']>30) & (diabetes['Age (years)']<=40)
Filtro_compreso3 = (diabetes['Age (years)']>40) & (diabetes['Age (years)']<=50)
Filtro_maggiore50 = diabetes['Age (years)'] >50

diabetes[Filtro_compreso3]                                                                  #per visualizzare il filtro (come risultato = tabella



# • Qual è la media della pressione sanguigna diastolica per le diverse fasce di età? 6 
#'Diastolic blood pressure (mm Hg)'

filtro_20 = diabetes.loc[Filtro_minore20, 'Diastolic blood pressure (mm Hg)']  
filtro_2030 = diabetes.loc[Filtro_compreso1, 'Diastolic blood pressure (mm Hg)']  
filtro_3040 = diabetes.loc[Filtro_compreso2, 'Diastolic blood pressure (mm Hg)']  
filtro_4050 = diabetes.loc[Filtro_compreso3, 'Diastolic blood pressure (mm Hg)']  
filtro_50 = diabetes.loc[Filtro_maggiore50, 'Diastolic blood pressure (mm Hg)']  
                                                                                             #mi creo un filtro che mi trasforma i valori booleani in valori nominativi e mi crea due tabelle


print('media sotto i 20 anni',filtro_20.mean())    #non vuole il round perché è NULL
print('media sotto i 30',filtro_2030.mean().round(2))
print('media sotto i 40',filtro_3040.mean().round(2))
print('media sotto i 50',filtro_4050.mean().round(2))
print('media dopo i 50',filtro_50.mean().round(2))
                                                                                              #al filtro ci calcolo la media r la stampo

# MANDARE TUTTO SEPARATAMENTE


# • Qual è la media della pressione per ogni singolo anno di età?

Filtro_eta = diabetes['Age (years)'] == 35                              #filtro che mi rilascia il true false sulle età che corrispondono a 35
filtro = diabetes.loc[Filtro_eta, ['Age (years)']]                      #filtro che li trtasforma in valori nominativi e mi crea due colonnine
print(filtro.mean().round(2))                                           #stampo il filtro con la media



```

```python
# DATABASE INSURANCE

insurance = pd.read_csv('DATABASES/insurance.csv')
insurance

# • Visualizziamone le dimensioni, un'anteprima, e osserviamo i nomi di colonna; 
# • Quali sono le medie di charges rispetto a region? Ci sono differenze significative? 
# • E rispetto a smoker? E a sex? 
# • Quali sono i descrittori statistici di bmi? Quali sono minimo, media e massimo di charges rispetto ai diversi quartili dei valori di bmi?

# • Visualizziamone le dimensioni, un'anteprima, e osserviamo i nomi di colonna; 
insurance.shape         #dimensioni
insurance.columns       #colonne
insurance
```

```python
# • Quali sono le medie di charges rispetto a region? Ci sono differenze significative? 

nomi_unici= insurance['region'].unique                            #usato per vedere i valori unici nella colonna region 
print(nomi_unici)
# southeast, southwest, northeast, northwest


SH = insurance['region'] == 'southeast'
SW = insurance['region'] == 'southwest'                           #ho creato dei filtri che mi fanno risultare TRUE solo i nomi che cerco per ogni filtro  == che corrispondono
NH = insurance['region'] == 'northeast'
NW = insurance['region'] == 'northwest'

mediaSH = insurance.loc[ SH, 'charges'].mean().round(2)
mediaSW = insurance.loc[ SW, 'charges'].mean().round(2)
mediaNH = insurance.loc[ NH, 'charges'].mean().round(2)                    #ho creato una tabella con le medie di ogni filtro sulla colonna charges (valore nominatovo e numerico, non più TRUE/FALSE)
mediaNW = insurance.loc[ NW, 'charges'].mean().round(2)

print('\n')

print('media in southeast:',mediaSH)
print('media in southwest:',mediaSW)
print('media in northeast:',mediaNH)                              #STAMPATO
print('media in northeast:',mediaNW)
```

```python
# • E rispetto a smoker? E a sex? media di charges per smokers e sex

filtro_smoker = insurance['smoker'] == 'yes'
filtro_smok = insurance['smoker'] == 'no'      

filtro_sex = insurance['sex'] == 'female'
fil_sex = insurance['sex'] == 'male'

media1 = insurance.loc[filtro_smoker, 'charges'].mean().round(2)
media2 = insurance.loc[filtro_smok, 'charges'].mean().round(2)
media3 = insurance.loc[filtro_sex, 'charges'].mean().round(2)
media4 = insurance.loc[fil_sex, 'charges'].mean().round(2)

print('la media di charges per fumatori:', media1)
print('la media di charges per non fumatori:',media2)
print('la media di charges per donne:',media3)
print('la media di charges per uomini:',media4)

# media = insurance.loc[[filtro_smoker, filtro_smok, filtro_sex, fil_sex, 'charges']].mean().round(2)      
```
